Improting modules and downloading dataset

```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
import matplotlib.pyplot as plt

tf.set_random_seed(1)

mnist = input_data.read_data_sets('./mnist', one_hot=True)

trainIndexNumber = [0, 12345, 23456, 34567, -100]
testIndexNumber = [0, 9876, 8765, 6543, -1]
```

Displaying images from training set

```python
for number in range(len(trainIndexNumber)):
    
    plt.imshow(mnist.train.images[trainIndexNumber[number]].reshape((28, 28)), cmap='gray')
    plt.show()
```

Displaying images from test set

```python
for number in range(len(testIndexNumber)):
    
    plt.imshow(mnist.test.images[testIndexNumber[number]].reshape((28, 28)), cmap='gray')
    plt.show()
```
